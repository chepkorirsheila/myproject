```python
#Add interactive widgets (sliders, dropdowns)
# Sidebar widgets
st.sidebar.title("Filters")

# 1. Year range slider
min_year = int(df['year'].min())
max_year = int(df['year'].max())
selected_years = st.sidebar.slider(
    "Select publication year range:",
    min_year,
    max_year,
    (min_year, max_year)
)

# Filter dataframe by year range
df_filtered = df[(df['year'] >= selected_years[0]) & (df['year'] <= selected_years[1])]

# 2. Journal dropdown
journals = ['All'] + sorted(df_filtered['journal'].dropna().unique().tolist())
selected_journal = st.sidebar.selectbox("Select a journal:", journals)

if selected_journal != 'All':
    df_filtered = df_filtered[df_filtered['journal'] == selected_journal]

# 3. Source dropdown
sources = ['All'] + sorted(df_filtered['source_x'].dropna().unique().tolist())
selected_source = st.sidebar.selectbox("Select a source:", sources)

if selected_source != 'All':
    df_filtered = df_filtered[df_filtered['source_x'] == selected_source]

# Display filters
st.write(f"Selected years: {selected_years[0]} - {selected_years[1]}")
st.write(f"Selected journal: {selected_journal}")
st.write(f"Selected source: {selected_source}")

st.write("Filtered dataset preview:")
st.dataframe(df_filtered.head())
```

```python
#Display your visualizations in the app
# Sidebar: interactive widgets
st.sidebar.title("Filters")

# Year range slider
min_year = int(df['year'].min())
max_year = int(df['year'].max())
selected_years = st.sidebar.slider(
    "Select publication year range:",
    min_year,
    max_year,
    (min_year, max_year)
)

# Filter by year
df_filtered = df[(df['year'] >= selected_years[0]) & (df['year'] <= selected_years[1])]

# Journal dropdown
journals = ['All'] + sorted(df_filtered['journal'].dropna().unique().tolist())
selected_journal = st.sidebar.selectbox("Select a journal:", journals)
if selected_journal != 'All':
    df_filtered = df_filtered[df_filtered['journal'] == selected_journal]

# Source dropdown
sources = ['All'] + sorted(df_filtered['source_x'].dropna().unique().tolist())
selected_source = st.sidebar.selectbox("Select a source:", sources)
if selected_source != 'All':
    df_filtered = df_filtered[df_filtered['source_x'] == selected_source]

# Main layout
st.title("CORD-19 Paper Analysis Dashboard")
st.write(f"Papers from {selected_years[0]} to {selected_years[1]}")
if selected_journal != 'All':
    st.write(f"Filtered by journal: {selected_journal}")
if selected_source != 'All':
    st.write(f"Filtered by source: {selected_source}")

# Line chart: publications over time
st.subheader("Number of Publications Over Time")
pub_per_year = df_filtered['year'].value_counts().sort_index()
fig1, ax1 = plt.subplots(figsize=(10,5))
pub_per_year.plot(kind='line', marker='o', color='blue', ax=ax1)
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Publications')
ax1.grid(True)
st.pyplot(fig1)

# Bar chart: top journals
st.subheader("Top Publishing Journals")
df_journal = df_filtered.dropna(subset=['journal'])
top_journals = df_journal['journal'].value_counts().head(10)
fig2, ax2 = plt.subplots(figsize=(10,5))
top_journals.plot(kind='bar', color='green', ax=ax2)
ax2.set_xlabel('Journal')
ax2.set_ylabel('Number of Publications')
ax2.set_xticklabels(top_journals.index, rotation=45)
st.pyplot(fig2)

# Horizontal bar chart: top sources
st.subheader("Top Sources")
df_source = df_filtered.dropna(subset=['source_x'])
source_counts = df_source['source_x'].value_counts().head(10)
fig3, ax3 = plt.subplots(figsize=(10,5))
source_counts.plot(kind='barh', color='orange', ax=ax3)
ax3.set_xlabel('Number of Papers')
ax3.set_ylabel('Source')
st.pyplot(fig3)

# Word cloud of paper titles
st.subheader("Word Cloud of Paper Titles")
titles = df_filtered['title'].dropna()
text = " ".join(titles)

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    stopwords=STOPWORDS,
    collocations=False
).generate(text)

fig4, ax4 = plt.subplots(figsize=(15,5))
ax4.imshow(wordcloud, interpolation='bilinear')
ax4.axis('off')
st.pyplot(fig4)
```

1. Overview

The CORD-19 dataset is a large collection of research papers related to COVID-19 and coronaviruses, including publications from journals, preprint servers, and repositories. This analysis focuses on trends in publications, top journals, sources, and frequent topics from paper titles.

2. Publications Over Time

Publications started increasing significantly in 2020, coinciding with the COVID-19 outbreak.

The line chart shows a steep rise in papers in early 2020, peaking around mid-2020.

This reflects the rapid surge of research to understand the virus, epidemiology, and clinical responses.
3. Top Publishing Journals

The top journals contributing to COVID-19 research include:

Journal of Infectious Diseases

medRxiv (preprint server)

bioRxiv (preprint server)

Journal of Public Health

4. Distribution by Source

The dataset includes papers from several sources: PMC, MEDRXIV, BIORXIV, and others.

PMC (PubMed Central) contributes the majority of publications, followed by MEDRXIV and BIORXIV.

The horizontal bar chart highlights which repositories are the primary contributors to the dataset.

5. Word Cloud of Paper Titles

The word cloud reveals common research topics:

Terms like “COVID”, “SARS-CoV-2”, “infection”, “vaccine”, “patients”, “transmission” appear most frequently.

This indicates that the majority of papers focus on:

Viral transmission and epidemiology

Clinical characteristics and patient outcomes

Prevention, vaccines, and treatments

6. Key Insights

Rapid growth in COVID-19 research started in early 2020.

Preprint servers are essential for fast communication of research.

Publications are concentrated in a few journals and sources, but the dataset covers global research outputs.

Word cloud analysis shows high focus on infection dynamics, vaccines, and public health interventions.
