```python
import streamlit as st
import pandas as pd
```

```python
%%writefile health_app.py
st.title("Chronic Condition Monitoring App")
#Inputing patient data
name = st.text_input("Patient Name")
blood_sugar = st.number_input("Blood Sugar Level(mg/dL)")
systolic = st.number_input("Systolic Blood Pressure")
diastolic = st.number_input("Diastolic Blood Pressure")
heart_rate = st.number_input("Heart Rate")
#Save button
if st.button("Save Health Record"):
  data = {
      "Name": [name],
      "Blood Sugar": [blood_sugar],
      "Systolic": [systolic],
      "Diastolic": [diastolic],
      "Heart Rate": [heart_rate]
  }
  df = pd.DataFrame(data)
  st.write("Health Record Saved!")
  st.write(df)
```
