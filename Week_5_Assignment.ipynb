Assignment 1: Design Your Own Class! üèóÔ∏è

1.Create a class representing anything you like (a Smartphone, Book, or even a Superhero!).

2.Add attributes and methods to bring the class to life!

3.Use constructors to initialize each object with unique values.

4.Add an inheritance layer to explore polymorphism or encapsulation

```python
# Base class representing a generic shoe
class Shoe:
    def __init__(self, brand, size, color):
        self.brand = brand
        self.size = size
        self.color = color

    def wear(self):
        print(f"Wearing {self.color} {self.brand} shoes of size {self.size}.")

    def info(self):
        print(f"Brand: {self.brand}, Size: {self.size}, Color: {self.color}")


# Child class representing a running shoe (inheritance + polymorphism)
class RunningShoe(Shoe):
    def __init__(self, brand, size, color, cushioning):
        super().__init__(brand, size, color)
        self.cushioning = cushioning

    # Polymorphism: override the wear method
    def wear(self):
        print(f"Running in {self.brand} shoes with {self.cushioning} cushioning.")

    # Additional method specific to running shoes
    def run_info(self):
        print(f"Brand: {self.brand}, Size: {self.size}, Color: {self.color}, Cushioning: {self.cushioning}")


# Creating objects
shoe1 = Shoe("Nike", 10, "black")
shoe2 = RunningShoe("Adidas", 9, "white", "extra")

# Using methods
shoe1.wear()
shoe1.info()
print("---")
shoe2.wear()
shoe2.run_info()
```

Activity 2: Polymorphism Challenge!

Create a program that includes animals or vehicles with the same action (like move()). However, make each class define move() differently (for example, Car.move() prints "Driving" , while Plane.move() prints "Flying" ).

```python
# Base class
class Vehicle:
    def move(self):
        pass  # Placeholder method, overridden in child classes


# Child class 1
class Car(Vehicle):
    def move(self):
        print("Driving ")


# Child class 2
class Plane(Vehicle):
    def move(self):
        print("Flying ")


# Child class 3
class Boat(Vehicle):
    def move(self):
        print("Sailing ")


# Create objects
my_car = Car()
my_plane = Plane()
my_boat = Boat()

# Polymorphism in action
vehicles = [my_car, my_plane, my_boat]

for v in vehicles:
    v.move()  # Each object calls its own version of move()
```
