1.)Create a program that reads a file and writes a modified version to a new file.


```python

# Open the original file for reading
with open("input.txt", "r", encoding="utf-8") as infile:
    lines = infile.readlines()

# Open a new file for writing the modified content
with open("output.txt", "w", encoding="utf-8") as outfile:
    for i, line in enumerate(lines, start=1):
        modified_line = f"{i}: {line.upper()}"
        outfile.write(modified_line)

print("File has been read and modified content written to output.txt")
```

2.)Ask the user for a filename and handle errors if it doesn’t exist or can’t be read.

```python

filename = input("Enter the filename to read: ")

try:
    infile = open(filename, "r")
    outfile = open("output.txt", "w")

    for line in infile:
        outfile.write(line)

    infile.close()
    outfile.close()
    print("File copied successfully.")

except FileNotFoundError:
    print("Error: File does not exist.")

except PermissionError:
    print("Error: File cannot be read.")
```
